## DNN

```python
#-*- coding: utf-8 -*-
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import Dropout, Dense
from keras.callbacks import ModelCheckpoint,EarlyStopping
import tensorflow as tf
import os
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
from tensorflow.keras.optimizers import *
from tensorflow.keras.callbacks import *
```

```python
# 딥러닝 모델 선언
model = Sequential()
model.add(Dense(256, input_dim=8, activation='elu'))  # 입력층
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(128, activation='elu'))      # 은닉층1 
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(64, activation='elu'))      # 은닉층1
model.add(Dense(32, activation='elu'))
model.add(Dense(16, activation='elu'))
model.add(Dense(8, activation='elu'))
model.add(Dense(8, activation='elu'))
model.add(Dense(1))# 출력층
# 선형 회귀는 마지막에 참과 거짓을 구분할 필요가 없음. 출력층에 활성화 함수를 지정할 필요도 없음

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop', 
              metrics=['acc'])
```

```python
# 모델 저장 폴더 만들기
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)

modelpath = "./model/{epoch:02d}-{val_loss:.4f}.hdf5"

# 모델 업데이트 및 저장
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_acc', verbose=0, save_best_only=True, mode='max')
# 콜백함수
reduce_lr = ReduceLROnPlateau(monitor='val_acc', factor=0.02, patience=100, mode='max')

# 학습 자동 중단 설정
early_stopping_callback = EarlyStopping(monitor='val_acc', patience=130, mode='max')
```

```python
history = model.fit(x, y, validation_split=0.3, epochs=200, batch_size=32, verbose=1, callbacks=[early_stopping_callback,checkpointer,reduce_lr])
print("\ntrain acc: %.4f" % (model.evaluate(x, y)[1]))               
```

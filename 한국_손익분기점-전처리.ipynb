## 감독 전처리

```python
df['directors'].nunique() #총 427명의 감독이 존재
```

```python
df['directors'].value_counts()[:100]
```

```python
director_list = df['directors'].value_counts()[:100] # 영화 2개 미만의 영화를 만든 감독은 기타 처리하는 함수 만들기
def cont_func(director):
    if director in director_list:
        return director
    else:
        return '기타'
```

```python
df['directors'] = df['directors'].apply(lambda x : cont_func(x)) #함수적용
```

```python
tr_direct_rank = df.groupby('directors').audiAcc.mean().reset_index(name = 'direct_rank').sort_values(by = 'direct_rank', ascending=False)
tr_direct_rank
# 감독들의 영화 관객수를 평균값으로 감독들의 순위 매김, 중앙값보다 의미 있는 결과를 보여서 평균값으로 함
# 실제로 유명한 감독들이 상위권에 있는 것을 확인할 수 있음
```

```python
tr_direct_rank[:70]
```

```python
tr_direct_rank['direct_rank'] = [i + 1 for i in range(tr_direct_rank.shape[0])] #순위 적용
tr_direct_rank
```

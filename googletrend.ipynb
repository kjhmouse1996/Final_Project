```python
import pandas as pd
import numpy as np
import pytrends
from pytrends.request import TrendReq
import datetime
import csv
```

```python
kw_list = []
```

```python
movie2 = pd.read_csv('final3.csv')
```

```python
movie2.columns
```

```python
date1 = movie2['개봉일']
```

```python
date2 = movie2['개봉일2']
```

```python
key = movie2['title']
```

```python
key.head()
```

```python
i = 0
```

```python
for i in range(4000,6393):
    kw_list = key[i:i+1]
    pytrends = TrendReq(hl = "ko", tz = 360)
    pytrends.build_payload(kw_list, cat=0, timeframe= "{0} {1}".format(date2[i], date1[i]), geo='', gprop='')
    getdatainfo = pytrends.interest_over_time()
    print(getdatainfo)
    getdatainfo.to_csv("trend%d.csv" % i, encoding="utf-8")
```

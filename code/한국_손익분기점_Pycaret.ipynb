# 모델링

## train_test_split

```python
df_fin = pd.read_csv('전처리_final.csv')
```

```python
df_fin
```

```python
y = df_fin['Target']
x = df_fin.drop('Target', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
 
sc = StandardScaler()
sc.fit(x)
 
x_std = sc.transform(x)
```

```python
df_std = pd.DataFrame(x_std)
```

```python
df_std['Target'] = y
```

```python
df_std
```

## pycaret

```python
from pycaret.classification import *
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_std, test_size=0.3, random_state=100)
```

```python
setup_clf = setup(data = train, target= 'Target', fold_shuffle=True)
```

```python
top10 = compare_models(sort='Accuracy', n_select=10, fold=5)
```

```python
tuned_top10 = [tune_model(i, fold=5) for i in top10[:4]]
```

```python
blender_top10 = blend_models(estimator_list=tuned_top10)
```

```python
final_model = finalize_model(blender_top10)
prediction = predict_model(final_model, data=test)
```

```python
prediction['Target'] = prediction['Target'].astype(int)
```

```python
prediction['Label'] = prediction['Label'].astype(float)
```

```python
prediction['Label'] = prediction['Label'].astype(int)
```

```python
from pycaret.utils import check_metric
check_metric(prediction['Target'], prediction['Label'], metric = 'Accuracy') 
#0.68 naver movie buzz 미포함 # 0.7225 네이버무비버즈 포함
# 구글트렌드 포함 0.7399
# 구글트렌드+ 네이버 합침 0.7341
# 0.7803 모든 피처사용
# 0.7919 4개, 0.7861 5개act_n 뺏을 경우
# 0.7803 5개, dist_rank 뺏을 경우
# 0.7861 4개, genre_rank 뺏을 경우
# 0.7341 4개, 0.7283 5개 direct_rank 뺏을 경우
# 0.7861 5개, 0.7803개 네이버 기대지수 뺏을 경우
# 0.7861 4개 star_buzz 뺏을 경우
# 0.763 4개, 0.7283 5개 naver_movie_buzz
# 0.7399 3개 google_trend
```

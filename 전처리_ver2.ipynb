# kmdb불러오기

```python
df_kmdb = pd.read_csv('KMDB_ALL_AFTER.csv')
```

```python
df_kmdb
```

## kmdb 전처리

```python
df_kmdb = df_kmdb.drop(['Unnamed: 0', 'kmdbUrl','screenCnt','salesAcc','audiAcc'], axis=1)
```

```python
df_kmdb
```

```python
df_kmdb.columns
```

```python
df_kmdb.isnull().sum()
```

```python
df_kmdb= df_kmdb.drop(df_kmdb[df_kmdb['prodYear'] < 2010.0].index) # 2010년 이전 영화 삭제
```

```python
df_kmdb['title']=df_kmdb['title'].str.strip() # 제목에 있는 공백 없애주기
```

```python
df_kmdb = df_kmdb.drop(df_kmdb[df_kmdb['genre']=='에로'].index)
```

```python
df_kmdb.isnull().sum()
```
